```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import xlrd
```

```python
# Load the excel data using pandas into a data frame with a reference to sheet name
xl_file = pd.read_excel(open('data/UniversalBank.xlsx','rb'), sheetname='Data')
# Display the first 10 rows
xl_file.head(10)
```

```python
# Normalize the data a bit to make sure columns like zip code and id are not being used
customers = pd.DataFrame(xl_file[["Age","Experience", "Income", "Family", "CCAvg", "Education", "Mortgage","Personal Loan", "Securities Account", "CD Account", "Online", "CreditCard"]])
customers.head(10)
```

```python
# Training and validation split of our data
tsize = np.floor(0.60 * len(customers))
data_train = customers.iloc[:int(tsize),:]
data_validation = customers.iloc[int(tsize):,:]
```

```python
# Length of the training data (60%)
len(data_train)
```

```python
# Length of the validation data (40%)
len(data_validation)
```

```python
# Pivot table using training data with online as the column and CreditCard and Personal Loan as the rows
# The aggfunc passing in a list that contains len is meant to show the count of each value
training_pivot_table = pd.pivot_table(data_train, index=['CreditCard', 'Personal Loan'],columns=['Online'], aggfunc=[len])
training_pivot_table
```
